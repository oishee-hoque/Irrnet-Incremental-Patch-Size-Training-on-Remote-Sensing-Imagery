```python
def preprocess_data(img, mask, num_class):
    # img = img / 255.0  # Example normalization
    mask = tf.squeeze(mask, axis=-1)
    img = tf.transpose(img, (2, 0, 1))
    print(mask.shape)
    # mask = tf.one_hot(tf.cast(mask, tf.int32), depth=num_class)  # Example one-hot encoding
    # mask = tf.transpose(mask, (2, 0, 1))
    return {"pixel_values": img, "labels": mask}

def decode_img(img, color_mode, target_size=(256, 256)):
    # Convert the compressed string to a 3D uint8 tensor
    img = tf.image.decode_png(img, color_mode)
    # Use `convert_image_dtype` to convert to floats in the [0,1] range.
    if(color_mode == 3):
        img = tf.image.convert_image_dtype(img, tf.float32)
    # Resize the image to the target size
    img = tf.image.resize(img, target_size)
    return img


def process_path(file_path, color_mode='rgb',img_size=(256,256)):
    # Load the raw data from the file as a string
    img = tf.io.read_file(file_path)
    img = decode_img(img, color_mode=color_mode,target_size=img_size)
    return img

def trainGenerator(train_img_path, train_mask_path, batch_size, num_class, seed=None, img_size=(256,256)):

    # Create datasets for images and masks
    img_paths = tf.data.Dataset.list_files(train_img_path + '/*', seed=seed)
    mask_paths = tf.data.Dataset.list_files(train_mask_path + '/*', seed=seed)

    img_dataset = img_paths.map(lambda x: process_path(x, 3,img_size=img_size), num_parallel_calls=tf.data.AUTOTUNE)
    mask_dataset = mask_paths.map(lambda x: process_path(x, 1,img_size=img_size), num_parallel_calls=tf.data.AUTOTUNE)

    # Zip the image and mask datasets together
    dataset = tf.data.Dataset.zip((img_dataset, mask_dataset))

    # Now apply preprocess_data to each (img, mask) pair
    dataset = dataset.map(lambda img, mask: preprocess_data(img, mask, num_class), num_parallel_calls=tf.data.AUTOTUNE)

    # Continue with batching, repeating, etc.
    dataset = dataset.batch(batch_size).repeat().prefetch(tf.data.AUTOTUNE)

    return dataset
```

```python
train_img_path = "/scratch/gza5dr/IrrigationTypeDetection/Experinments/Data/train_images/train"
train_mask_path = "/scratch/gza5dr/IrrigationTypeDetection/Experinments/Data/train_masks/train"
train_img_gen = trainGenerator(train_img_path, train_mask_path,batch_size, num_class=3,seed=42,img_size=(IMG_HEIGHT,IMG_HEIGHT))

val_img_path = "/scratch/gza5dr/IrrigationTypeDetection/Experinments/Data/val_images/val"
val_mask_path = "/scratch/gza5dr/IrrigationTypeDetection/Experinments/Data/val_masks/val"
val_img_gen = trainGenerator(val_img_path, val_mask_path, batch_size, num_class=3,seed=42,img_size=(IMG_HEIGHT,IMG_HEIGHT))
```

```python
model_checkpoint = "nvidia/mit-b0"
id2label = {0: "other", 1: "sprinkler", 2: "flood"}
label2id = {label: id for id, label in id2label.items()}
num_labels = len(id2label)
model = TFSegformerForSemanticSegmentation.from_pretrained(
    model_checkpoint,
    num_labels=num_labels,
    id2label=id2label,
    label2id=label2id,
    ignore_mismatched_sizes=True,
)
```

```python
def data_prep(y_true, y_pred,num_classes=3):
    y_pred_classes = tf.argmax(y_pred, axis=1)
    # Upsample predictions to match y_truth's spatial dimensions
    y_pred_resized = tf.squeeze(tf.image.resize(y_pred_classes[..., tf.newaxis], (256, 256), method=tf.image.ResizeMethod.NEAREST_NEIGHBOR),axis=-1)
    # Ensure y_true is of an integer type for tf.one_hot
    y_true_int = tf.cast(y_true, tf.int32)

    # One-hot encode y_true and y_pred_classes
    y_true_one_hot = tf.one_hot(y_true_int, depth=num_classes)
    y_pred_one_hot = tf.one_hot(y_pred_resized, depth=num_classes)

    # Calculate TP, FP, FN for each class
    TP = tf.reduce_sum(tf.cast(y_true_one_hot * y_pred_one_hot, tf.float32), axis=[0, 1, 2])
    FP = tf.reduce_sum(tf.cast((1 - y_true_one_hot) * y_pred_one_hot, tf.float32), axis=[0, 1, 2])
    FN = tf.reduce_sum(tf.cast(y_true_one_hot * (1 - y_pred_one_hot), tf.float32), axis=[0, 1, 2])

    return TP, FP, FN

def iou(y_true, y_pred,num_classes=3):
    tp, fp, fn = data_prep(y_true, y_pred,num_classes=3)
    TP_total = tf.reduce_sum(tp)
    FP_total = tf.reduce_sum(fp)
    FN_total = tf.reduce_sum(fn)

    # Calculate IoU
    iou = TP_total / (TP_total + FP_total + FN_total + 1e-6)  # Adding a small epsilon to avoid division by zero

    return iou
def accuracy(y_true, y_pred,num_classes=3):
    y_pred_classes = tf.cast(tf.argmax(y_pred, axis=1), tf.int32)
    # Upsample predictions to match y_truth's spatial dimensions
    y_pred_resized = tf.squeeze(tf.image.resize(y_pred_classes[..., tf.newaxis], (256, 256), method=tf.image.ResizeMethod.NEAREST_NEIGHBOR),axis=-1)
    y_true_int = tf.cast(y_true, tf.int32)
    accuracy = tf.reduce_mean(tf.cast(tf.equal(y_true_int, y_pred_resized), tf.float32))

    return accuracy  
    
def precisions(y_true, y_pred,num_classes=3):
    tp, fp, fn = data_prep(y_true, y_pred,num_classes=3)
    precision = tp / (tp + fp + 1e-5)
    return precision
    
def recalls(y_true, y_pred,num_classes=3):
    tp, fp, fn = data_prep(y_true, y_pred,num_classes=3)
    recall = tp / (tp + fn + 1e-5)
    return recall
    
def f1_score(y_true, y_pred,num_classes=3):
    precision =  precisions(y_true, y_pred,num_classes=3)
    recall = recalls(y_true, y_pred,num_classes=3)
    f1_scores = 2 * ((precision * recall) / (precision + recall + 1e-5)) 
    return f1_scores
    

def precision_0(y_true, y_pred,num_classes=3):
    precision =  precisions(y_true, y_pred,num_classes=3)
    return precision[0]
    
def precision_1(y_true, y_pred,num_classes=3):
    precision =  precisions(y_true, y_pred,num_classes=3)
    return precision[1]
    
def precision_2(y_true, y_pred,num_classes=3):
    precision =  precisions(y_true, y_pred,num_classes=3)
    return precision[2]
    
def recall_0(y_true, y_pred,num_classes=3):
    recall = recalls(y_true, y_pred,num_classes=3)
    return recall[0]
    
def recall_1(y_true, y_pred,num_classes=3):
    recall = recalls(y_true, y_pred,num_classes=3)
    return recall[1]
    
def recall_2(y_true, y_pred,num_classes=3):
    recall = recalls(y_true, y_pred,num_classes=3)
    return recall[2]

    
def f1_score0(y_true, y_pred,num_classes=3):
    f1_scores=f1_score(y_true, y_pred,num_classes=3)
    return f1_scores[0]

    
def f1_score1(y_true, y_pred,num_classes=3):
    f1_scores=f1_score(y_true, y_pred,num_classes=3)
    return f1_scores[1]

    
def f1_score2(y_true, y_pred,num_classes=3):
    f1_scores=f1_score(y_true, y_pred,num_classes=3)
    return f1_scores[2]
```
